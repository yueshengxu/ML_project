```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
l = pickle.load( open( "result.pkl", "rb" ) )
l_no_fair = pickle.load( open( "result_no_fair.pkl", "rb" ) )
l_xndcg = pickle.load( open( "result_xndcg.pkl", "rb" ) )
l_xndcg10 = pickle.load( open( "result_xndcg10.pkl", "rb" ) )
l_ratio1 = pickle.load( open( "result_ratio0.1.pkl", "rb" ) )

epochs = np.arange(len(l))
```

```python
train_loss_ratio1 = []
fair_loss_ratio1 = []
ndcg_ratio1 = []
ndcg1_ratio1 = []
ndcg0_ratio1 = []
for i in l_ratio1:
    train_loss_ratio1.append(i[0].item())
    fair_loss_ratio1.append(i[1].item())
    ndcg_ratio1.append(i[2])
    ndcg1_ratio1.append(i[3])
    ndcg0_ratio1.append(i[4])
    
fair_loss_no_fair = []
train_loss_no_fair = []
ndcg_no_fair = []
ndcg1_no_fair = []
ndcg0_no_fair = []
for i in l_no_fair:
    train_loss_no_fair.append(i[0].item())
    fair_loss_no_fair.append(i[1].item())
    ndcg_no_fair.append(i[2])
    ndcg1_no_fair.append(i[3])
    ndcg0_no_fair.append(i[4])
    
fair_loss_xndcg = []
train_loss_xndcg = []
ndcg_xndcg   = []
ndcg1_xndcg  = []
ndcg0_xndcg  = []
for i in l_xndcg:
    train_loss_xndcg.append(i[0].item())
    fair_loss_xndcg.append(i[1].item())
    ndcg_xndcg.append(i[2])
    ndcg1_xndcg.append(i[3])
    ndcg0_xndcg.append(i[4])
    
```

```python
plt.plot(epochs,train_loss_no_fair,color = 'red',label = 'no fair')
plt.plot(epochs,train_loss_ratio1,color = 'green',label = 'group ratio')
plt.plot(epochs,train_loss_xndcg,color = 'blue',label = 'absolute difference')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Average Ranking Loss')
plt.title('Training Loss vs. Epoch')
```

```python
plt.plot(epochs,fair_loss_ratio1,color = 'green')
plt.plot(epochs,fair_loss_no_fair,color = 'red')
plt.plot(epochs,fair_loss_xndcg,color = 'yellow')
```

```python
plt.plot(epochs,ndcg_no_fair,color = 'red',label='no fairness')
plt.plot(epochs,ndcg_ratio1,color = 'green',label='group ratio')
plt.plot(epochs,ndcg_xndcg,color = 'blue',label='absolute difference')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('True NDCG Score')
plt.title('Testing True NDCG vs. Epoch')


```

```python
plt.plot(epochs,np.array(ndcg1_no_fair) - np.array(ndcg0_no_fair),color = 'red',label='no fairness')
plt.plot(epochs,np.array(ndcg1_ratio1)-np.array( ndcg0_ratio1),color = 'green',label='group ratio')
plt.plot(epochs,np.array(ndcg1_xndcg)-np.array( ndcg0_xndcg),color = 'blue',label='absolute difference')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('True NDCG Score')
plt.title('Testing Group 1 True NDCG vs. Epoch')
```

```python
plt.plot(epochs,ndcg1_no_fair,color = 'red',label='no fairness 1')
plt.plot(epochs,ndcg1_ratio1,color = 'green',label='group ratio 1')
plt.plot(epochs,ndcg1_xndcg,color = 'blue',label='absolute difference 1')
```
